```python
#sepreating the features and target data
x=df.drop('salary',axis=1)
x.head()
```

```python
y=df['salary']
y
```

```python
#checking the skewness in our feature data
x.skew()
```

```python
#removing the skewness
from sklearn.preprocessing import power_transform
input=power_transform(x)
x=pd.DataFrame(power_transform(x),columns=x.columns)
x.head()
```

```python
x.shape
```

```python
y.shape
y=y.values.reshape(-1,1)
```

```python
y.shape
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit_transform(x)
x=sc.fit_transform(x)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
#selecting the best Random state
maxAccu=0
maxRS=0
for i in range (0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    lr=LogisticRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc > maxAccu:
        maxAccu=acc
        maxRs=i
print ('best Accuracy Score is :',maxAccu, 'on random_state:',maxRS)
    
```

```python
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier 
dt=DecisionTreeClassifier ()
dt.fit(x_train,y_train)
pred=dt.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
pred=rf.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
pred=svc.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

#here we are geting heighest accuracy with RandomForestClassifier i.e 61 percent
